```python
strength = [183,188,198,198,201,205,209,210,210,211,215,219,225,234,235,238,241,245,251,253,255,259,261,261,262,285,289,295,303,305,311,312,315,315,317,317,321,321,321,321,324,328,331,333,341,342,348,348,349,352]
thickness = [20,25,15,22,18,11,10,18,20,15,17,20,35,30,25,21,25,28,24,30,25,15,18,25,20,30,23,28,30,20,24,27,25,22,32,25,33,29,30,20,40,36,55,26,28,33,28,25,31,27]

bim_data = [(l,w) for l, w in zip(strength,thickness)]
bim_target = [1] * 26 + [0] * 24

import numpy as np
input_array = np.array(bim_data)
target_array = np.array(bim_target)

np.random.seed(45)
index = np.arange(50)
np.random.shuffle(index)


train_input = input_array[index[:37]]
train_target = target_array[index[:37]]

test_input = input_array[index[37:]]
test_target = target_array[index[37:]]

print(train_input, train_target)
print(test_input, test_target)


import matplotlib.pyplot as plt
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(test_input[:,0], test_input[:,1])
plt.xlabel('strength')
plt.ylabel('thickness')
plt.show()
```
