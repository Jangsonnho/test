# Generator

```python
def method():
    print("Start method()")
    val=[]
    for x in range(0,5):
        val.append(x**2)
    return val
```

```python
out = method()
for i in range(0,5):
    print(out[i])
```

```python
def generator():
    print("Start generator()")
    for x in range(0,5):
        yield x**2

gen = generator()
for i in range(0,5):
    print(next(gen))
```

#데이터 타입마다 다른 for loop 스타일

```python
items = [[1,2],[3,4],[5,6]]
for item in items:
    print(item[0], item[1])
for item1, item2 in items:
    print(item1, item2)
```

```python
info = {'A':1, 'B':2, 'C':3}
for key in info:
    print(key, info[key])
for key, value in info.items():
    print(key,value)
```

## 2.3.2 zip가 들어간 for loop

```python
items1 = [[1,2], [3,4], [5,6]]
items2 = [['A','B'], ['C','D'], ['E','F']]
print(items1)
print(items2)
```

```python
for digits, characters in zip(items1, items2):
    print(digits, characters)
```

```python
import matplotlib.pyplot as plt

x = np.arange(0.,5.,0.2)
y = x **2
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0.,5.,1.125)
y = x **2
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0.,5.,0.2)
y1 = x
y2 = 0.25 * x**2
y3 = np.cos(x)
plt.plot(x,y1, '-')
plt.plot(x,y2, '--')
plt.plot(x,y3, ':')
plt.legend(['y=x', 'y=0.25* x**2', 'y= cos(x)'])
plt.show()
```

```python
x = np.arange(0.,5.,0.2)
y1 = x
y2 = 0.25 * x**2
y3 = np.cos(x)

plt.plot(x,y1,'o-',x,y2,'--s',x,y3,':x')
plt.title('Plot Tutorial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['$y=x$', '$y=0.25 * x^2$', '$y=cos(x)$'])
plt.show()
```

```python
plt.plot(x,y1,'o-', label =r'$y=x$')
plt.plot(x,y2,'--s', label=r'$y=\frac{1}{4}x^2$')
plt.plot(x,y3,':x', label=r'$y=cos(x)$')
plt.title('Plot Tutorial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
plt.subplot(3,1,1)
plt.plot(x,y1,'-ok')
plt.title("Subplots Example")
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,1,2)
plt.plot(x,y2,'--sk')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,1,3)
plt.plot(x,y3,":xk")
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show
```
