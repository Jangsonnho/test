```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

$$
f(x) = x^2 - 4x + 6
$$

```python
f = lambda x : x**2 - 4*x +6 
x= np.linspace(-1,6,100)
y = f(x)

# 함수 그래프 그리기
plt.plot(x,y,'k-')
plt.plot(2,2,'sk') # 최소값 위치
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
grad_f = lambda x : 2*x -4 # 그래디언트

x0 =0.0 #초기값
MaxIter = 10 # 반복횟수
learning_rate = 0.25 # 학습률
print("step\tx\tf(x)")
print("{:02d}\t{:6.5f}\t{:6.5f}".format(0,x0,f(x0)))

for i in range(MaxIter):
    x1 = x0 - learning_rate * grad_f(x0)
    x0 = x1 #업데이트
    print("{:02d}\t{:6.5f}\t{:6.5f}".format(i+1, x0, f(x0)))
```

```python
def steepest_decent(func, grad_func, x0, learning_rate=0.01, MaxIter=10, verbose=True):
    paths = []
    for i in range(MaxIter):
        x1 = x0 -learning_rate * grad_func(x0)
        if verbose:
            print('{0:03d} : {1:4.3f}, {2:4.2E}'.format(i, x1, func(x1)))
        x0 = x1
        paths.append(x0)
    return(x0, func(x0), np.array(paths))

xopt, fopt, paths = steepest_decent(f, grad_f, 0.0, learning_rate=0.2, MaxIter=9)

plt.plot(x,y,'-k')
plt.plot(2,2,'sk-')
plt.plot(paths, f(paths), 'ko:')
for k, point in enumerate(paths):
    plt.text(point, f(point), '$x_{0}$'.format(k),verticallalignment='bottom')

plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
